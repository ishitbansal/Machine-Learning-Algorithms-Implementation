```python
import pandas as pd
import numpy as np
np.set_printoptions(threshold=np.inf)
```

```python
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
```

```python
class HierarchicalClustering:
    def linkages(self, X, linkage_type):
        linked = linkage(X, method=linkage_type)
        return linked

    def dendrogram(self, Z):
        dendrogram(Z, orientation='top', distance_sort='descending',
                   show_leaf_counts=True)
        plt.title('Hierarchical Clustering Dendrogram')
        plt.xlabel('Sample Index')
        plt.ylabel('Distance')
        plt.show()
```

```python
data = pd.read_csv('SMAI-Dataset-hc-dataset/new_customers.csv')
```

```python
print(data.dtypes)
```

```python
data = data.drop(columns=['CustomerID'])
data['Genre'] = data['Genre'].replace({'Male': 1, 'Female': 0})
print(data)
```

```python
hc = HierarchicalClustering()
```

The choice of linkage and features can significantly affect the clustering results

Single Linkage - This method calculates the shortest distance between clusters. It tends to form elongated clusters and is sensitive to outliers.

Complete Linkage - This method calculates the longest distance between clusters. It tends to form compact, spherical clusters.

Average Linkage - This method calculates the average distance between all pairs of points in two clusters. It balances between the effects of single and complete linkage.

Centroid Linkage - This method calculates the distance between the centroids of two clusters. It can produce well-balanced clusters.

Ward's Linkage - This method minimizes the increase in variance when two clusters merge. It tends to produce compact, roughly equally sized clusters.

Here, we observe that single linkage forms elongated clusters, complete linkages form compact clusters, average linkage somewhat balances single and complete linkages, centroid linkage also forms well balanced clusters and ward linkage froms relatively equal sized clusters

```python
subset_features = [['Genre', 'Annual Income (k$)'], ['Age', 'Spending Score (1-100)'], [
    'Genre', 'Spending Score (1-100)'], ['Age', 'Annual Income (k$)']]
linkage_types = ['single', 'complete', 'average', 'centroid', 'ward']
for feature in subset_features:
    d = data[feature]
    for l in linkage_types:
        print('Features:', feature)
        print('Linkage:', l)
        linked = hc.linkages(d, l)
        hc.dendrogram(linked)
```

Here, we observe that single linkage forms elongated clusters, complete linkages form compact clusters, average linkage somewhat balances single and complete linkages, centroid linkage also forms well balanced clusters and ward linkage froms relatively equal sized clusters
